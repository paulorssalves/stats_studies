```r
miles_traveled = c(89, 66, 78, 111, 44, 77, 80, 66, 109, 76)
num_deliveries = c(4, 1, 3, 6, 1, 3, 3, 2, 5, 3)
gas_price = c(3.84, 3.19, 3.78, 3.89, 3.57, 3.57, 3.03, 3.51, 3.54, 3.25)
travel_time = c(7, 5.4, 6.6, 7.4, 4.8, 6.4, 7, 5.6, 7.3, 6.4)

d = lm(travel_time ~ miles_traveled)
```

```r
sm = summary(d)
```

```r
plot(travel_time ~ miles_traveled)
abline(d)
sm
```

```r
sm$sigma
```

```r
sm$sigma
# sigma is the standard error of the regression
```

```r
tn = lm (travel_time ~num_deliveries)
```

```r
tn
```

```r
plot(travel_time ~ num_deliveries)
abline(tn)
```

```r
k = summary(tn)
k$sigma
```
